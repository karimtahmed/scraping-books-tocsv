```python
response = requests.request('GET', 'https://books.toscrape.com/index.html')
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
all_articles = soup.find_all('article', attrs={'class': 'product_pod'})
```

```python
def page2():   
    div1=soup.find('div', attrs={'class': 'side_categories'}).find('ul')
    cats=div1.find_all('a')
    lis2=[]
    for cat in cats:
        for i in range(len(cats)):

            categloge=cat.get('href')
        lis2.append(categloge)
    return lis2
```

```python
def path():   
    div1=soup.find('div', attrs={'class': 'side_categories'}).find('ul')
    cats=div1.find_all('a')
    lis3=[]
    for cat in cats:
        c=cat.get_text().strip().split('/')
        lis3.append(c)
    return lis3
```

```python

listOfFunc=page2()
listOfFunc
```

```python

def get_books(page):
    s=0
    response = requests.request('GET', page)
    soup = BeautifulSoup(response.content, 'html.parser')
    all_articles = soup.find_all('article', attrs={'class': 'product_pod'})
    rates = {'One': 1, 'Two': 2, 'Three': 3, 'Four': 4, 'Five': 5}
    s=s+1
    for article in all_articles:
        s=s+1
        price = float(article.find('p', attrs={'class': 'price_color'}).get_text().split('£')[1])
        book_name = article.find('h3').find('a').get('title')
        rating = article.find('p').get('class')[1]
        sql="insert into book (name,rate,price,cat_id) values(?,?,?,?)"
        data =(book_name, rates[rating], price,s)
        cursor=connection.execute(sql,data)
        connection.commit()
        print(f'Book name: {book_name}\nPrice: {price}\nRate: {rates[rating]}\n------')
```
